```python
# reset index
shop_df= shop_df.reset_index(drop=True)
```

```python
# Create a linear regression model between open price and the difference of open and peak price
# Prepare test and training datasets for leniar regression
from sklearn import linear_model
from sklearn.model_selection import train_test_split


X = np.array(shop_df[['Open']])
Y = np.array(shop_df[['Diff_High_Open']])

# 80% Training and 20% testing
x_train, x_test, y_train, y_test = train_test_split(X, Y, train_size=0.8)

# Train the model
regr = linear_model.LinearRegression()
regr.fit(x_train, y_train)

coeff = regr.coef_
intercpt = regr.intercept_

# print values
print("coefficient", coeff)
print("Intercept", intercpt)
```

```python
# Predict the outcome
y_pred = regr.predict(x_test)
```

```python
# Compare the actual values with the predicted values
shop_regr = pd.DataFrame({'Actual': y_test.flatten(), 'Predicted': y_pred.flatten()})
shop_regr
```

```python
shop_regr_df= shop_regr.head(50)
shop_regr_df.plot(kind='bar',figsize=(16,10))
plt.plot(which='major', linestyle='-', linewidth='0.8', color='green')
plt.plot(which='minor', linestyle=':', linewidth='0.8', color='black')
plt.title("Predicted profit vs. Actual profit When stocks are bought at any opening price")
plt.show()
```

```python
# Find Mean values
from sklearn import metrics
print('Mean Absolute Error:', metrics.mean_absolute_error(y_test, y_pred))  
print('Mean Squared Error:', metrics.mean_squared_error(y_test, y_pred))  
print('Root Mean Squared Error:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
```
